```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
```

```python
from transformers import pipeline

filler = pipeline("fill-mask")
result = filler("veda is a <mask> good girl")
```

```python
print(result)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification", model="bert-base-cased")
labels = ["education", "dance", "tvshow"]
re = classifier("This is a course about the Transformers library", labels)
```
